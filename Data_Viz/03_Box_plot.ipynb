```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
kashti = sns.load_dataset("titanic")
sns.boxplot(x="survived",
           y="age",showmeans=True,
             meanprops={"marker": "*",
                         "markersize": "12",
                         "markeredgecolor": "red"},
           data=kashti)
plt.xlabel("How many survived", size="10", weight='bold'),
plt.ylabel("Age(years)", size="10", weight='bold'),
plt.title("Box plot of survive or dead", size="14", weight='bold')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# FacetGrid using catplot to make multiple box plots by sex
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker",
    col="sex",         # facet by 'sex' → creates a separate plot for Male and Female
    kind="box",
    height=4,
    aspect=1
)

plt.suptitle("Box Plots of Total Bill by Day, Split by Sex", y=1.05)
plt.show()
```
