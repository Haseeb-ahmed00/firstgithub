### Grouping

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
kashti = sns.load_dataset("titanic")
kashti
# draw a line plot
sns.barplot(x="sex", y="alone",hue="who",  data=kashti)
plt.show()
```

### Orders

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
kashti = sns.load_dataset("titanic")
kashti
# draw a line plot
sns.barplot(x="sex", y="alone",hue="who",  data=kashti, order=["female","male"])
plt.show()
```

### Show different color

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
kashti = sns.load_dataset("titanic")
kashti
# draw a line plot
sns.barplot(x="sex", y="alone",hue="who",  data=kashti,  palette = "Reds")
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
kashti = sns.load_dataset("titanic")
kashti
# draw a line plot
sns.barplot(x="sex", y="alone",hue="who",  data=kashti, color = "Red", errorbar=None,
            palette='pastel')
plt.show()
```

```python
#import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt
#load dataset
kashti = sns.load_dataset("titanic")
kashti
# draw a line plot
sns.barplot(x="class",y="fare",hue="sex",  data=kashti, estimator=mean)
plt.show()
```

```python
#import libraries
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt
#load dataset
kashti = sns.load_dataset("titanic")
kashti
# draw a line plot
sns.barplot(x="class",y="fare",hue="sex",  data=kashti, estimator=median, saturation=0.1)
plt.show()
```

```python
#horizontal plot
#import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt
#load dataset
kashti = sns.load_dataset("titanic")
kashti
# draw a line plot
sns.barplot(x="fare",y="class",hue="sex",  data=kashti, estimator=mean, saturation=1)
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# read titanic .csv file
# form seaborn library
kashti = sns.load_dataset("titanic")

sns.barplot(x="class", y="fare", data=kashti,
           linewidth=2.5, facecolor=(1,1,1,0),
           errcolor="0.2", edgecolor="0.2")
```
